```python
def space_cleaner(dato):
    r = re.compile(r"^\s*|\s*$", re.IGNORECASE)
    dato = r.sub(r"", dato)
    return dato
```

### Socios
- juntar nombre + apellido para tener el nombre completo
- agregar columna sid
- sid sera PK

```python
socios = pd.read_csv("Datos Proyecto - Socios.csv")
for col in socios.columns.tolist():
    socios[col] = socios[col].apply(space_cleaner)
socios["sid"] = socios.index
socios["sid"] = socios.sid.apply(lambda x: x+1)
socios = socios[["sid", "nombre", "apellido", "nacionalidad"]]
socios["nombre_completo"] = socios.nombre + " " + socios.apellido
```

### Proyectos
- columna PID será agregada a la tabla
- PID seran PK
- Comuna y Region se agregaran a la tabla Lugares y se agregará el LID del lugar a la tabla de relacion PID Y LID
- separar Socios y ver que no se repitan
- agregar a la tabla de relacion PID y SID por cada Socio
- crear las tablas IsA ProyectoMinera, ProyectoCentral y ProyectoVertedero, con los respectivos PID y
    mineral o energia, (puede ayudar la columna tipo, 1:Minera, 2:Central, 3:Vertedero)
- eliminar columna Socio
- eliminar columna Comuna y Region
- eliminar columna tipo, mineral, tipo generacion

```python
proyectos = pd.read_csv("Datos Proyecto - Proyectos.csv")
for col in proyectos.columns.tolist()[1:]:
    proyectos[col] = proyectos[col].apply(space_cleaner)
proyectos["pid"] = proyectos.index
proyectos["pid"] = proyectos.pid.apply(lambda x: x+1)
cols = proyectos.columns.tolist()
proyectos = proyectos[cols[-1:]+cols[:-1]]
```

```python
def poner_tipo(n):
    if n == 1:
        return "Minera"
    elif n == 2:
        return "Central"
    elif n == 3:
        return "Vertedero"
```

```python
Lugares = pd.DataFrame([], columns=["lid","comuna", "region"])
PyL = pd.DataFrame([], columns=["pid", "lid"])
PyS = pd.DataFrame([], columns=["pid", "sid"])
ProyectoCentral = pd.DataFrame([], columns=["pid", "energia"])
ProyectoVertedero = pd.DataFrame([], columns=["pid"])
ProyectoMinera = pd.DataFrame([], columns=["pid"])
PyM = pd.DataFrame([], columns=["pid", "mid"])
minerales = pd.DataFrame([], columns=["mid", "nombre"])
```

```python
pbar = ProgressBarCounter(proyectos.shape[0])
for indx in proyectos.index:
    comuna = proyectos.at[indx, "comuna"]
    r = re.compile(r"^\s|\s$", re.IGNORECASE)
    comuna = r.sub(r"", comuna)
    if not (Lugares.comuna == comuna).any():
        Lugares = pd.concat([Lugares, pd.DataFrame([[0, comuna, proyectos.at[indx, "región"]]],
                                                                   columns=["lid","comuna", "region"])],
                                                                                            axis=0, ignore_index=True)
        lid = Lugares.query("comuna == \"{}\"".format(comuna)).index[0] + 1
        Lugares.at[lid-1, "lid"] = lid
    else:
        lid = Lugares.query("comuna == \"{}\"".format(comuna)).lid.values[0]
    PyL = pd.concat([PyL, pd.DataFrame([[proyectos.at[indx, "pid"], lid]], columns=["pid", "lid"]) ],
                                                                                        axis=0, ignore_index=True)
    socios_p = set(proyectos.at[indx, "socios"].split(","))
    for socio in socios_p:
        r = re.compile(r"^\s*|\s*$", re.IGNORECASE)
        socio = r.sub(r"", socio)
        if not (socios.nombre_completo == socio).any():
            print("EL SOCIO NO EXISTE CTM!")
        else:
            sid = socios.query("nombre_completo == \"{}\"".format(socio)).sid.values[0]
            PyS = pd.concat([PyS, pd.DataFrame([[proyectos.at[indx, "pid"], sid]], columns=["pid", "sid"]) ],
                                                                                    axis=0, ignore_index = True)
    tipo = proyectos.at[indx, "tipo"]
    if tipo == 1:
        ProyectoMinera = pd.concat([
                                ProyectoMinera,
                                pd.DataFrame([[proyectos.at[indx, "pid"]]], columns=["pid"]) 
                                    ], axis=0, ignore_index = True)
        minerales_p = set(proyectos.at[indx, "mineral"].split(","))
        for mineral in minerales_p:
            r = re.compile(r"^\s*|\s*$", re.IGNORECASE)
            mineral = r.sub(r"", mineral)
            if not (minerales.nombre == mineral).any():
                minerales = pd.concat([minerales, pd.DataFrame([[0, mineral]], columns=["mid","nombre"])],
                                                                    axis=0, ignore_index = True)
                mid = minerales.query("nombre == \"{}\"".format(mineral)).index[0] + 1
                minerales.at[mid-1, "mid"] = mid
            else:
                mid = minerales.query("nombre == \"{}\"".format(mineral)).mid.values[0]
            PyM = pd.concat([PyM, pd.DataFrame([[proyectos.at[indx, "pid"], mid]], columns=["pid", "mid"]) ],
                                                                                    axis=0, ignore_index = True)
    elif tipo == 2:
        energia = proyectos.at[indx, "tipo generación"]
        r = re.compile(r"^\s*|\s*$", re.IGNORECASE)
        energia = r.sub(r"", energia)
        ProyectoCentral = pd.concat([
                                ProyectoCentral,
                                pd.DataFrame([[proyectos.at[indx, "pid"], energia]], columns=["pid", "energia"]) 
                                    ], axis=0, ignore_index = True)
    elif tipo == 3:
        ProyectoVertedero = pd.concat([
                                ProyectoVertedero,
                                pd.DataFrame([[proyectos.at[indx, "pid"]]], columns=["pid"]) 
                                    ], axis=0, ignore_index = True)
    else:
        print("ERROR: EL PROYECTO NO ES NADA!")
    pbar.notify()
```

```python
socios = socios[["sid", "nombre", "apellido", "nacionalidad"]]
proyectos["tipo"] = proyectos.tipo.apply(poner_tipo)
proyectos_out = proyectos[['pid', 'nombre', 'tipo','latitud', 'longitud', 'fecha apertura', 'operativo']]
proyectos_out.columns = ["pid", "nombre", "tipo", "latitud", "longitud", "fecha_apertura", "operativo"]
```

### Recursos
- columna RID será agregada a la tabla
- RID y Número seran PK
- separar ONGs y ver que no se repitan
- agregar a la tabla de relacion RID y OID por cada ONG
- pasar el nombre a pid del proyecto y agregarlo a la tabla de relacion RID y PID
- Comuna y Region se agregaran a la tabla Lugares y se agregará el LID del lugar a la tabla de relacion RID Y LID
- crear las tablas IsA RecursosAprobados, RecursosRechazados y RecursosPendientes, con los respectivos RID, status y
    fecha_dictamen
- cambiar las fechas al format dd-mm-yyyy
- eliminar la columna ONG
- eliminar las columnas status y fecha_dictamen
- Eliminar la columna Comuna y Región

```python
recursos = pd.read_csv("Datos Proyecto - Recursos.csv")
for col in ['número',
 'nombre',
 'causa contaminante',
 'área influencia (kms)',
 'descripción',
 'fecha apertura',
 'región de tramitación',
 'comuna de tramitación',
 'status',
#  'fecha_dictamen',
 'ONGs que participan']:
    recursos[col] = recursos[col].apply(space_cleaner)
recursos["fecha apertura"] = recursos["fecha apertura"].apply(lambda x: x.replace("/","-"))
recursos["rid"] = recursos.index
recursos["rid"] = recursos.rid.apply(lambda x: x+1)
c = recursos.columns.tolist()
recursos = recursos[c[-1:] + c[:-1]]
```

```python
RyO = pd.DataFrame([], columns=["rid", "oid"])
ONGs = pd.DataFrame([], columns=["oid", "nombre"])
PyR = pd.DataFrame([], columns=["pid", "rid"])
RyL = pd.DataFrame([], columns=["rid", "lid"])
RecursosAprobados = pd.DataFrame([], columns=["rid", "fecha_dictamen"])
RecursosRechazados = pd.DataFrame([], columns=["rid", "fecha_dictamen"])
RecursosPendientes = pd.DataFrame([], columns=["rid"])
```

```python
pbar = ProgressBarCounter(recursos.shape[0])
for indx in recursos.index:
    if recursos.at[indx, "ONGs que participan"]!="#VALUE!":
        ongs = set(recursos.at[indx, "ONGs que participan"].split("/"))
        for ong in ongs:
            r = re.compile(r"^\s*|\s*$", re.IGNORECASE)
            ong = r.sub(r"", ong)
            if not (ONGs.nombre == ong).any():
                ONGs = pd.concat([ONGs, pd.DataFrame([[0, ong]], columns=["oid", "nombre"])],
                                                                    axis=0, ignore_index = True)
                oid = ONGs.query("nombre == \"{}\"".format(ong)).index[0] + 1
                ONGs.at[oid-1, "oid"] = oid
            else:
                oid = ONGs.query("nombre == \"{}\"".format(ong)).oid.values[0]
            RyO = pd.concat([RyO, pd.DataFrame([[recursos.at[indx, "rid"], oid]], columns=["rid", "oid"]) ],
                                                                                    axis=0, ignore_index = True)
    nombre_proyecto = recursos.at[indx, "nombre"].replace("Recurso contra ", "")
    r = re.compile(r"^\s*|\s*$", re.IGNORECASE)
    nombre_proyecto = r.sub(r"", nombre_proyecto)
    pid = proyectos.query("nombre == \"{}\"".format(nombre_proyecto)).pid.values[0]
    PyR = pd.concat([PyR, pd.DataFrame([[pid, recursos.at[indx, "rid"]]], columns=["pid", "rid"]) ],
                                                                                axis=0, ignore_index = True)
    comuna = recursos.at[indx, "comuna de tramitación"]
    r = re.compile(r"^\s|\s$", re.IGNORECASE)
    comuna = r.sub(r"", comuna)
    if not (Lugares.comuna == comuna).any():
        Lugares = pd.concat([Lugares, pd.DataFrame([[0, comuna, recursos.at[indx, "región de tramitación"]]],
                                                                   columns=["lid","comuna", "region"])],
                                                                                            axis=0, ignore_index=True)
        lid = Lugares.query("comuna == \"{}\"".format(comuna)).index[0] + 1
        Lugares.at[lid-1, "lid"] = lid
    else:
        lid = Lugares.query("comuna == \"{}\"".format(comuna)).lid.values[0]
    RyL = pd.concat([RyL, pd.DataFrame([[recursos.at[indx, "rid"], lid]], columns=["rid", "lid"]) ],
                                                                                        axis=0, ignore_index=True)
    status = recursos.at[indx, "status"]
    r = re.compile(r"^\s*|\s*$", re.IGNORECASE)
    status = r.sub(r"", status)
    if status == "aprobado":
        RecursosAprobados = pd.concat([
                                RecursosAprobados,
                                pd.DataFrame([[recursos.at[indx, "rid"], recursos.at[indx, "fecha_dictamen"]]],
                                                                                         columns=["rid", "fecha_dictamen"]) 
                                    ], axis=0, ignore_index = True)
    elif status == "rechazado":
        RecursosRechazados = pd.concat([
                                RecursosRechazados,
                                pd.DataFrame([[recursos.at[indx, "rid"], recursos.at[indx, "fecha_dictamen"]]],
                                                                                         columns=["rid", "fecha_dictamen"]) 
                                    ], axis=0, ignore_index = True)
    elif status == "en trámite":
        RecursosPendientes = pd.concat([
                                RecursosPendientes,
                                pd.DataFrame([[recursos.at[indx, "rid"]]], columns=["rid"]) 
                                    ], axis=0, ignore_index = True)
    else:
        print("ERROR: EL RECURSO NO ES NADA!")
    pbar.notify()
```

```python
recursos_out = recursos[['rid', 'número', 'nombre', 'causa contaminante',
                         'área influencia (kms)', 'descripción', 'fecha apertura', 'status']]
recursos_out.columns = ["rid", "numero", "nombre", "causa_contaminante", "area_influencia_kms", "descripcion", "fecha_apertura", "status"]
```
